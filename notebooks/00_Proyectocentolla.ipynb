```python
import pandas as pd

archivos = [
    '2019_0304_desembarque_total_por_puerto.xlsx',
    '2020_0304_desembarque_total_por_puerto.xlsx',
    '2021_0304_desembarque_total_por_puerto.xlsx',
    '2022_0304_desembarque_total_por_puerto.xlsx',
    '2023_0304_desembarque_total_por_puerto.xlsx',
    '2024_0304_Desembarque-total-por-puerto.xlsx',
]

dfs = []  # Lista para guardar los DataFrames de cada año

for archivo in archivos:
    print(f"\nCargando {archivo} ...")
    df = pd.read_excel(f'data/raw/{archivo}', header=4)
    df['Año'] = archivo[:4]  # Extrae el año del nombre del archivo
    dfs.append(df)
    print(df.head(3))  # Muestra un avance de cada DataFrame cargado
```

```python
# Lista de nombres de meses esperados (puede ajustar si ves variantes)
meses = ['ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']

dfs_limpios = []

for df in dfs:
    # Si el primer nombre de columna no es "ESPECIE", lo renombro
    col_0 = df.columns[0]
    if not col_0.upper().startswith('ESPECIE'):
        df = df.rename(columns={col_0: 'ESPECIE'})

    # Renombro las columnas por posición si es necesario
    columnas_nuevas = ['ESPECIE'] + meses + ['Total', 'Año']
    df.columns = columnas_nuevas[:len(df.columns)]  # Ajusta si hay columnas extra

    # Elimino filas vacías y sólo dejo las que tienen especie no nula
    df = df[df['ESPECIE'].notnull()]
    # Quito espacios en blanco en los nombres de especie
    df['ESPECIE'] = df['ESPECIE'].str.strip().str.upper()

    # Me quedo solo con la Centolla
    df_centolla = df[df['ESPECIE'].str.contains('CENTOLLA', na=False)]

    dfs_limpios.append(df_centolla)

# Uno todos los años en un solo DataFrame
df_todo_centolla = pd.concat(dfs_limpios, ignore_index=True)

# Veo un resumen final
df_todo_centolla.head(5)
```

# 📑 Cargar los datasets de esfuerzo pesquero, temperatura y regulaciones

Leo los archivos CSV/XLSX de datos procesados que voy a integrar al dataset principal.

```python
import pandas as pd

# Cargar dataset de esfuerzo pesquero
esfuerzo = pd.read_csv('data/raw/esfuerzo_pesquero_punta_arenas_2019_2024.csv')

# Cargar dataset de temperatura superficial del mar (nombre puede variar)
# Si es Excel, usá read_excel
temperatura = pd.read_csv('data/raw/E20200016_18062025174150841_temperaturapuntaarena.csv')

# Cargar regulaciones
import pandas as pd

# Rango de años según tus datos
anios = range(2019, 2025)
meses = range(1, 13)
reg_data = []

for anio in anios:
    for mes in meses:
        veda = 1 if 2 <= mes <= 11 else 0
        reg_data.append({'Año': anio, 'Mes': mes, 'veda_centolla': veda})

regulaciones = pd.DataFrame(reg_data)
print(regulaciones.head(15))

# Verificamos las primeras filas de cada uno
print(esfuerzo.head())
print(temperatura.head())
print(regulaciones.head())
```

# 🌡️ Limpieza del archivo de temperatura superficial del mar

Dejo sólo los registros del período 2019-2024, extraigo año y mes en número, y me quedo con la temperatura promedio mensual.

```python
import pandas as pd

# Cargar temperatura
temperatura = pd.read_csv('data/raw/E20200016_18062025174150841_temperaturapuntaarena.csv')

# Extraigo año y mes en número de la columna 'Mes'
temperatura['Año'] = temperatura['Mes'].str.extract(r'(\d{4})').astype(int)
# Diccionario para pasar de texto a número de mes
meses_map = {
    'Enero': 1, 'Febrero': 2, 'Marzo': 3, 'Abril': 4, 'Mayo': 5, 'Junio': 6,
    'Julio': 7, 'Agosto': 8, 'Septiembre': 9, 'Octubre': 10, 'Noviembre': 11, 'Diciembre': 12
}
temperatura['Mes_num'] = temperatura['Mes'].str.extract(r'(\w+)$')[0].map(meses_map)
# Filtrar sólo 2019-2024 y quedarnos con las columnas clave
temperatura = temperatura[(temperatura['Año'] >= 2019) & (temperatura['Año'] <= 2024)]
temperatura = temperatura[['Año', 'Mes_num', 'Value']]
temperatura = temperatura.rename(columns={'Mes_num': 'Mes', 'Value': 'temp_promedio'})

print(temperatura.head())
```

# 🔗 Integración final de todos los datasets: Centolla, Esfuerzo Pesquero, Temperatura y Regulaciones

En este paso:
- Paso el DataFrame de Centolla a formato largo (cada fila = Año, Mes, Volumen).
- Hago merge progresivo con esfuerzo pesquero, temperatura y regulaciones por Año y Mes.
- El resultado es un dataset listo para modelar y análisis.

```python
# Paso 1: Derretir (melt) el DataFrame de centolla
id_vars = ['ESPECIE', 'Año', 'Total']
value_vars = ['ENE', 'FEB', 'MAR', 'ABR', 'MAY', 'JUN', 'JUL', 'AGO', 'SEP', 'OCT', 'NOV', 'DIC']
df_centolla_largo = df_todo_centolla.melt(id_vars=id_vars, value_vars=value_vars,
                                          var_name='Mes', value_name='volumen_centolla')

# Paso 2: Pasar nombre de mes a número
mes_map = {
    'ENE': 1, 'FEB': 2, 'MAR': 3, 'ABR': 4, 'MAY': 5, 'JUN': 6,
    'JUL': 7, 'AGO': 8, 'SEP': 9, 'OCT': 10, 'NOV': 11, 'DIC': 12
}
df_centolla_largo['Mes'] = df_centolla_largo['Mes'].map(mes_map)
df_centolla_largo['Año'] = df_centolla_largo['Año'].astype(int)
df_centolla_largo['Mes'] = df_centolla_largo['Mes'].astype(int)

# Paso 3: Merge con esfuerzo pesquero
df_full = df_centolla_largo.merge(esfuerzo, on=['Año', 'Mes'], how='left')

# Paso 4: Merge con temperatura
df_full = df_full.merge(temperatura, on=['Año', 'Mes'], how='left')

# Paso 5: Merge con regulaciones
df_full = df_full.merge(regulaciones, on=['Año', 'Mes'], how='left')

# Resultado final
print(df_full.head(10))
```
